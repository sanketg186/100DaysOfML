## K-Nearest Neighbor
In this notebook we will implement K nearest neighbor from scratch

```python
class KNearestNeighbor:
    def __init__(self,X,y,K=5):
        self.X = X
        self.y = y
        self.K = K
    
    def l2_dist(self,Xtest):
        dists = np.sum(self.X**2,axis=1) + np.sum(Xtest*Xtest,axis=1)[None].T - 2*np.dot(Xtest,self.X.T) 
        return dists
    
    def predict(self,X):
        dists =self.l2_dist(X)
        self.y_pred = np.zeros(X.shape[0])
        for i in range(dists.shape[0]):
            y_values = np.argsort(dists[i, :])
            k_closest_classes = self.y[y_values[: self.K]].astype(int)
            self.y_pred[i] = np.argmax(np.bincount(k_closest_classes))

        return self.y_pred
```

```python
random_seed = 22
X, y = make_blobs(n_samples=200, centers=2,random_state=random_seed)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=random_seed)
```

```python
knn = KNearestNeighbor(X_train, y_train,K=5)
print("Train accuracy")
y_pred = knn.predict(X_train)
print(f"Accuracy: {sum(y_pred == y_train) / y_train.shape[0]}")
print("-------------------------------------------")
print("Test accuracy")
y_pred = knn.predict(X_test)
print(f"Accuracy: {sum(y_pred == y_test) / y_test.shape[0]}")
```
