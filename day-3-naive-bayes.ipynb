## Day 3: Naive Bayes classifier
In this notebook we will implement a naive Bayes classfier

```python
class NaiveBayes():
    def prior(self,c):
        prior_prob = np.mean(self.y==c)
        return prior_prob
    
    def likelihood(self,mu,var,x):
        eps = 1e-4
        coef = 1.0 / np.sqrt(2.0 * np.pi * var + eps)
        exponent = np.exp(-(np.power(x - mu, 2) / (2 * var + eps)))
        likelihood_prob = coef*exponent
        return likelihood_prob
    
    def fit(self,X,y):
        self.x = X
        self.y = y
        self.classes = np.unique(y)
        
        self.parameters=[]
        
        for i, c in enumerate(self.classes):
            x_given_c = X[np.where(y == c)]
            self.parameters.append([])
            for col in x_given_c.T:
                parameter = {"mean": col.mean(), "var": col.var()}
                self.parameters[i].append(parameter)

    def classify(self,x):
        posteriors = []
        for i, c in enumerate(self.classes):
            posterior_prob = self.prior(c)
            for feature_value, params in zip(x, self.parameters[i]):
                likelihood_prob = self.likelihood(params["mean"], params["var"], feature_value)
                posterior_prob *= likelihood_prob
            posteriors.append(posterior_prob)
        # Return the class with the largest posterior probability
        return self.classes[np.argmax(posteriors)]
    
    def predict(self,X):
        y_pred = np.array([self.classify(sample) for sample in X])
        return y_pred
        
        
```

```python
random_seed = 21
X, y = make_blobs(n_samples=1000, centers=2,random_state=random_seed)
naive_bayes = NaiveBayes()
naive_bayes.fit(X, y)
```

```python
y_pred= naive_bayes.predict(X)
```

```python
print("Accuracy: ", np.mean(y==y_pred))
```
